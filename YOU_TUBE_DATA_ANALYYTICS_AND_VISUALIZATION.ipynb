```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 channels by average views
top_channels = df.groupby('channel_title')['view_count'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_channels.values, y=top_channels.index)
plt.title('Top 10 Channels by Average Views')
plt.xlabel('Average Views')
plt.ylabel('Channel Title')
plt.tight_layout()
plt.show()
```

```python
# Convert published_date to datetime
df['published_date'] = pd.to_datetime(df['published_date'])
df['published_date_only'] = df['published_date'].dt.date
```

```python
import matplotlib.pyplot as plt

# Count videos by date
video_counts = df.groupby('published_date_only').size()

# Plot
plt.figure(figsize=(12,6))
video_counts.plot()
plt.title('Number of Videos Published Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Videos')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot total views by date
views_over_time = df.groupby('published_date_only')['view_count'].sum()

plt.figure(figsize=(12,6))
views_over_time.plot(color='green')
plt.title('Total Views Over Time')
plt.xlabel('Date')
plt.ylabel('Total Views')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Top 5 most active channels
top_channels = df['channel_title'].value_counts().head(5).index

# Filter data
filtered_df = df[df['channel_title'].isin(top_channels)]

# Plot
plt.figure(figsize=(12,6))
for channel in top_channels:
    data = filtered_df[filtered_df['channel_title'] == channel]
    daily_uploads = data.groupby('published_date_only').size()
    plt.plot(daily_uploads, label=channel)

plt.title('Upload Activity of Top 5 Channels')
plt.xlabel('Date')
plt.ylabel('Number of Videos')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Total likes per day
likes_over_time = df.groupby('published_date_only')['like_count'].sum()

plt.figure(figsize=(12,6))
likes_over_time.plot(color='orange')
plt.title('Total Likes Over Time')
plt.xlabel('Date')
plt.ylabel('Total Likes')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Total comments per day
comments_over_time = df.groupby('published_date_only')['comment_count'].sum()

plt.figure(figsize=(12,6))
comments_over_time.plot(color='red')
plt.title('Total Comments Over Time')
plt.xlabel('Date')
plt.ylabel('Total Comments')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# If category_id has known mappings, you can map them here (optional)
category_names = {
    1: 'Film & Animation', 2: 'Autos & Vehicles', 10: 'Music', 15: 'Pets & Animals',
    17: 'Sports', 20: 'Gaming', 22: 'People & Blogs', 23: 'Comedy', 24: 'Entertainment',
    25: 'News & Politics', 26: 'Howto & Style', 27: 'Education', 28: 'Science & Technology',
    29: 'Nonprofits & Activism'
}
df['category_name'] = df['category_id'].map(category_names)

# Total views by category
category_views = df.groupby('category_name')['view_count'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
category_views.plot(kind='bar', color='purple')
plt.title('Total Views by Category')
plt.xlabel('Category')
plt.ylabel('Total Views')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np  # For scaling values

# Your existing top channels (means)
top_channels = df.groupby('channel_title')['view_count'].mean().sort_values(ascending=False).head(10)

# Create temp df for violin: scale means to simulate data distribution
temp_data = []
for channel, mean_views in top_channels.items():
    # Repeat values around mean (e.g., 100 points with noise for density)
    views = np.random.normal(mean_views, mean_views * 0.2, 100)  # 20% std dev
    for v in views:
        temp_data.append({'channel_title': channel, 'view_count': max(0, v)})  # No negatives

temp_df = pd.DataFrame(temp_data)

# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=temp_df, x='view_count', y='channel_title', inner='quartile')
plt.title('Top 10 Channels by Average Views (Distribution)')
plt.xlabel('Average Views')
plt.ylabel('Channel Title')
plt.tight_layout()
plt.show()
```
