```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
list1 = [[1,2,np.nan,4],[np.nan,11,6,9]]
df= DataFrame(list1,index=['row1','row2'],columns=['col1','col2','col3','col4'])
df
```

```python
df.sum()    #default axis=0 -> on first axis(along rows)
```

```python
df.sum(axis=1)  #on the second axis
```

```python
df.min()   #default axis=0(along rows)
```

```python
df.idxmin()  #gives the index of the min value
```

```python
df.max(axis=1)  #along columns
```

MERGING

MERGING IS ONE OF THE METHODS USED TO JOIN DATAFRAME(MERGING,CONCATENATING,JOINING).

```python
df1 = DataFrame({'col1':np.arange(6),'common':['a','c','b','c','a','b']})
df1
```

```python
df2 = DataFrame({'col2':np.arange(1,5),'common':['a','m','b','c']})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='common')
```

```python
pd.merge(df1,df2,on='common',how='right')
```

```python
pd.merge(df1,df2,on='common',how='left')
```

```python
pd.merge(df1,df2,on='common',how='inner')
```

```python
pd.merge(df1,df2,on='common',how='outer')
```
